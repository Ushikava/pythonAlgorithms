```python
import numpy as np
import scipy.special as scipy
import matplotlib.pyplot
%matplotlib inline
```

```python
class neuralNetwork:
    
    def __init__(self, inputnodes, hiddennodes, outputnodes, learningrate):
        self.inodes = inputnodes
        self.hnodes = hiddennodes
        self.onodes = outputnodes
        
        #self.wih = (np.random.rand(self.hnodes, self.inodes) - 0.5)
        #self.who = (np.random.rand(self.onodes, self.hnodes) - 0.5)
        self.wih = np.random.normal(0.0, pow(self.hnodes, -0.5), (self.hnodes, self.inodes))
        self.who = np.random.normal(0.0, pow(self.onodes, -0.5), (self.onodes, self.hnodes))
        
        self.lr = learningrate
        
        # сигмоида в качестве функции активации
        self.activation_function = lambda x: scipy.expit(x)
        pass
    
    def train(self, inputs_list, targets_list):
        inputs = np.array(inputs_list, ndmin=2).T
        targets = np.array(targets_list, ndmin=2).T
        
        hidden_inputs = np.dot(self.wih, inputs)
        hidden_outputs = self.activation_function(hidden_inputs)
        
        final_inputs = np.dot(self.who, hidden_outputs)
        final_outputs = self.activation_function(final_inputs)
       
        # ошибки = целевое значение - фактическое значение
        output_errors = targets - final_outputs
        # скрытые ошибки = предыдущие, но распределенные по весовым коэффициентам связей
        hidden_errors = np.dot(self.who.T, output_errors)
        
        # обновление весовых коэффициентов связей между скрытым и выходным слоями
        self.who += self.lr * np.dot((output_errors * final_outputs * (1.0 - final_outputs)), np.transpose(hidden_outputs))
        # обновление весовых коэффициентов связей между входным и скрытым слоями
        self.wih += self.lr * np.dot((hidden_errors * hidden_outputs * (1.0 - hidden_outputs)), np.transpose(inputs))
        
        pass
    
    def query(self, inputs_list):
        #преобразование списка входных значений в двумерный массив
        inputs = np.array(inputs_list, ndmin=2).T
        
        hidden_inputs = np.dot(self.wih, inputs)
        hidden_outputs = self.activation_function(hidden_inputs)
        
        final_inputs = np.dot(self.who, hidden_outputs)
        final_outputs = self.activation_function(final_inputs)
        return final_outputs
    
```

```python
# количество узлов (входыне, скрытые и выходные)
input_nodes = 784
hidden_nodes = 200
output_nodes = 10

# коэффициент обучения
learning_rate = 0.1

test_network = neuralNetwork(input_nodes, hidden_nodes, output_nodes, learning_rate) 
```

```python
traning_data_file = open("mnist_train.csv", 'r')
traning_data_list = traning_data_file.readlines()
traning_data_file.close()
```

```python
epochs = 5

for e in range(epochs):
    for record in traning_data_list:
        all_values = record.split(",")

        inputs = (np.asfarray(all_values[1:])/255.0 * 0.99) + 0.01

        targets = np.zeros(output_nodes) + 0.01

        targets[int(all_values[0])] = 0.99
        test_network.train(inputs, targets)
        pass
    pass
```

```python
test_data_file = open("mnist_test.csv", 'r')
test_data_list = test_data_file.readlines()
test_data_file.close()
```

```python
scorecard = []

for record in test_data_list:
    all_values = record.split(',')
    
    correct_label = int(all_values[0])
    print(correct_label, "- по факту")
    
    inputs = (np.asfarray(all_values[1:])/255.0 * 0.99) + 0.01
    
    outputs = test_network.query(inputs)
    label = np.argmax(outputs)
    print(label, "- от нейросетки")
    if (label == correct_label):
        scorecard.append(1)
    else:
        scorecard.append(0)
        pass
    
    pass
```

```python
scorecard_array = np.asarray(scorecard)
print("Правильность ответов составила: ", (scorecard_array.sum() / scorecard_array.size))
```
